```python
m=data['Price_in_thousands']
```

```python
data.drop('Price_in_thousands',inplace=True,axis=1)
```

```python
data.insert(value=m,column='Price_in_thousands',loc=10)
```

```python
data.head()
```

```python
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.15,random_state=27)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
```

```python
linreg.fit(X_train,y_train)
```

```python
y_pred=linreg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
metrics.r2_score(y_test,y_pred)
```

```python
import pickle
file=open('car_price_prediction.pkl','wb')
pickle.dump(linreg,file)
```

```python
fm=open('car_price_prediction.pkl','rb')
reg=pickle.load(fm)
```

```python
y_pred=reg.predict(X_test)
```

```python
metrics.r2_score(y_test,y_pred)
```
